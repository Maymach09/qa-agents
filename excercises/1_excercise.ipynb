# Single Node Example

```python
from langgraph.graph import StateGraph, START, END
```

```python
from typing import Dict, TypedDict
```

# Add State

```python
class AgentState(TypedDict):
    user_input: str
```

# Add Node


```python
def compliment_node(state: AgentState) -> AgentState:
    """ This node is to compliment the person who's name will be provided in the input."""
    
    state['user_input'] = state['user_input'] + ', you are a genius!'

    return state
```

# build graph

```python
graph = StateGraph(AgentState)

graph.add_node("Compliment", compliment_node)

graph.set_entry_point("Compliment")
graph.set_finish_point("Compliment")

app = graph.compile()
```

```python
result = app.invoke({"user_input":"Mayank"})
```
