# 2. Single Node with multiple inputs

```python
from langgraph.graph import StateGraph
```

```python
from typing import Dict, TypedDict, List
```

```python
class AgentState(TypedDict):
    numbers: List[int]
    name: str
    operation: str
    response: str
```

# Node

```python
def compute_node(state: AgentState) -> AgentState:
    """ This node computes the sum or product of a list of numbers based on the operation specified in the state."""
    
    if state['operation'] == '+':
        result = sum(state['numbers'])
    elif state['operation'] == '*':
        result = 1
        for num in state['numbers']:
            result *= num
    else:
        result = 0  # Default case if operation is unknown

    state['response'] = f"Hello {state['name']}, your answer is: {result}."
    return state
```

# Graph

```python
graph = StateGraph(AgentState)
graph.add_node("Compute", compute_node)
graph.set_entry_point("Compute")
graph.set_finish_point("Compute")

app = graph.compile()
```

```python
result = app.invoke({"name":"Mayank", "numbers":['19','22','330',"4"], "operation":"*"})
result
```
